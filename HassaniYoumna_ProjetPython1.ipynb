 - *Étape 3* : **Création de nouvelles colonnes de calcul pour le calcul du VWAP**, qui est une moyenne pondérée où les poids sont les volumes (ici représentés par 'amount') et les valeurs sont les prix.

```python
# Ajout d'une colonne calculée pour chaque DataFrame
for df, suffix in zip(dataframes, suffixes):
    df[f'PjQj_{suffix}'] = df[f'price_{suffix}'] * df[f'amount_{suffix}'] #calcul de Pj*Qj (formule du sujet), nouvelle colonne ajoutee
    # qui représente le rpoduit du prix par la quantité (= volume) pour chaque transaction.
    print(f"Added PjQj for {suffix}: Top rows")
    print(df.head())  # Affiche les premières lignes pour vérifier nos calculs
    print(df.info())
```

**PARTIE 3 : CALCUL**

- *Étape 1* : **Agrégation des données pour le calcul de VWAP**

```python
# On cree un dictionnaire pour stocker les DataFrames agrégés
hourly_vwap_dataframes = {}

for df, suffix in zip(dataframes, suffixes):
    hourly_data = df.resample('H').agg({ # agrégation sur une base horaire
        f'PjQj_{suffix}': 'sum',
        f'amount_{suffix}': 'sum'
    })

# Creation de la colonne VWAP dans le DataFrame agrégé : calcul du VWAP pour chaque heure
    hourly_data[f'VWAP_{suffix}'] = hourly_data[f'PjQj_{suffix}'] / hourly_data[f'amount_{suffix}']
    hourly_vwap_only = hourly_data[[f'VWAP_{suffix}']] # creation d'un nouveau DataFrame avec uniquement VWAP

# Stockage du DataFrame dans le dictionnaire
    hourly_vwap_dataframes[suffix] = hourly_vwap_only

# Afficher les premières lignes des données agrégées pour vérifier
    print(f"Hourly VWAP for {suffix}:")
    print(hourly_vwap_only.head())
    print(f"Shape of hourly data for {suffix}: {hourly_vwap_only.shape}") # permet de verifier le nombre de lignes et de colonnes
    # comme c'est un volume horaire, nous devons obtenir 24 lignes en principe...

# Combinaison de tous les DataFrames VWAP horaires en un seul DataFrame
combined_hourly_vwap_df = pd.concat(hourly_vwap_dataframes.values(), axis=1)
```

- *Etape 2* : Calcul du Global VWAP horaire

```python
# Dictionnaire pour stocker les DataFrames agrégés avec les sommes nécessaires pour le calcul du VWAP global
global_vwap_dataframes = {}

for df, suffix in zip(dataframes, suffixes):
    hourly_data = df.resample('H').agg({
        f'PjQj_{suffix}': 'sum',
        f'amount_{suffix}': 'sum'
    })

    # Calcul du VWAP pour chaque heure et stockage des sommes nécessaires pour le VWAP global
    hourly_data[f'VWAP_{suffix}'] = hourly_data[f'PjQj_{suffix}'] / hourly_data[f'amount_{suffix}']
    global_vwap_dataframes[suffix] = hourly_data[[f'PjQj_{suffix}', f'amount_{suffix}']]

# Création d'un seul DataFrame à partir du dictionnaire
combined_hourly_df = pd.concat(global_vwap_dataframes.values(), axis=1, keys=suffixes)

# Calcul du VWAP global en utilisant les sommes des Pj*Qj et des montants de tous les exchanges
combined_hourly_df['Global_PjQj'] = combined_hourly_df.apply(lambda row: sum(row[x][f'PjQj_{x}'] for x in suffixes), axis=1)
combined_hourly_df['Global_amount'] = combined_hourly_df.apply(lambda row: sum(row[x][f'amount_{x}'] for x in suffixes), axis=1)
combined_hourly_df['Global_VWAP'] = combined_hourly_df['Global_PjQj'] / combined_hourly_df['Global_amount']

# Affichage des résultats : ce DataGrame contient plus qu'une colonne, on traite ce 'problème' juste après
print("Combined Hourly Global VWAP DataFrame:")
print(combined_hourly_df[['Global_VWAP']].head())
```

*Étape 2* : trier les Dataframes par ordre croissant

```python
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]
sorted_dataframes = []

for df, suffix in zip(dataframes, suffixes):
    # Trier chaque DataFrame par la colonne 'price_{suffix}' en ordre croissant
    sorted_df = df.sort_values(by=f'price_{suffix}', ascending=True)
    sorted_dataframes.append(sorted_df)

    # Afficher le DataFrame trié pour vérification
    print(sorted_df[[f'timestamp_{suffix}', f'price_{suffix}', f'amount_{suffix}']])
```

```python
# il nous faut créer une fonction à partir des explications
# du site web Wikipédia

def weighted_median(data, weights):
    # Assurez-vous que les données et les poids sont des tableaux numpy
    data, weights = np.array(data), np.array(weights)
    # Tri des données et des poids selon les valeurs des données
    sorted_indices = np.argsort(data)
    sorted_data = data[sorted_indices]
    sorted_weights = weights[sorted_indices]
    # Calcul de la somme cumulée des poids
    cumulative_weights = np.cumsum(sorted_weights)
    # Calcul de la médiane
    cutoff = cumulative_weights[-1] / 2.0
    # Retourne la valeur de la médiane pondérée
    return sorted_data[cumulative_weights >= cutoff][0]

# Application de la fonction sur l'ensemble des dataframes triés
for df, suffix in zip(sorted_dataframes, suffixes):
    median_price = weighted_median(df[f'price_{suffix}'], df[f'amount_{suffix}'])
    print(f"Weighted median for {suffix}: {median_price}")

# on a affiché les médianes pondérées pour chaque exchange
```

*Étape 3* : **Calcul de la Médiane Pondérée par heure pour chaque exchange**

```python

# Définition de la fonction de médiane pondérée à partir des explications du site web Wikipédia

def weighted_median(data, weights):
    data, weights = np.array(data), np.array(weights)
    if len(data) == 0: #si on a pas de données à traiter sur une heure donnée
        return np.nan  # Retourne NaN si les données sont vides
    sorted_indices = np.argsort(data)
    sorted_data = data[sorted_indices]
    sorted_weights = weights[sorted_indices]
    cumulative_weights = np.cumsum(sorted_weights)
    cutoff = cumulative_weights[-1] / 2.0
    return sorted_data[cumulative_weights >= cutoff][0]

# Chargement des données et traitement initial
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]
suffixes = ["stmp", "okcn", "lmax", "krkn", "itbi", "gmni", "cbse", "btrx", "bnus", "bfnx", "bfly"]
hourly_medians = {} # on crée un dictionnaire vide pour stocker les résultats

for df, suffix in zip(dataframes, suffixes):
    df.set_index(f'timestamp_{suffix}', inplace=True) # on indexe par les dates
    df.sort_values(by=f'price_{suffix}', inplace=True)
    # Calcul de la médiane pondérée pour chaque heure
    hourly_median = df.resample('H').apply(lambda x: weighted_median(x[f'price_{suffix}'], x[f'amount_{suffix}'])) # agrégation des données pour
    # le calcul du VWMP
    hourly_medians[suffix] = hourly_median.to_frame(name=f'Weighted_Median_{suffix}') # on transforme en dataframe

# Combinaison de tous les DataFrames en un seul pour faciliter l'analyse
combined_median_df = pd.concat(hourly_medians.values(), axis=1)

# Afficher le DataFrame combiné
print(combined_median_df)
```
